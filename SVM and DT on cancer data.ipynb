```python
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()
df['diagnosis'] = lbl.fit_transform(df['diagnosis'])
```

```python
plt.figure(figsize = (20,15))
crr = df.corr()
sns.heatmap(crr,annot = True,linewidth =2,fmt = '.2g')
#plt.tightlayout()
```

```python
X = df.drop('diagnosis',axis=1)
Y = df['diagnosis']
```

```python
plt.figure(figsize=(20,15))
plotnumber =1

for col in X.columns:
  if plotnumber <31:
    plt.subplot(6,5,plotnumber)
    sns.boxplot(X[col])
  plotnumber +=1

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state = 0,test_size = .2)
```

```python
from sklearn.preprocessing import StandardScaler
strd = StandardScaler()
x_train = strd.fit_transform(x_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf  = DecisionTreeClassifier(criterion = 'log_loss')
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
acc = accuracy_score(y_pred,y_test)
```

```python
y_tracc = clf.predict(x_train)
```

```python
acc_tran_dt = accuracy_score(y_tracc,y_train)
acc_tran_dt
```
